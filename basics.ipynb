# Module 5: Tensor Operations

Learn:

- Addition
- Subtraction

- Matrix Multiplication

```python
a = tf.constant([1,2,3])
b = tf.constant([4,5,6])

print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b)
print(a**b)
```

# Module 6: Broadcasting

One of the most important concepts.

Example:

```python
matrix = tf.constant([
    [1,2,3],
    [4,5,6]
])

scalar = 10

print(matrix + scalar)
```

# Module 8: GradientTape

Most important concept in Deep Learning.

Example:

```python
x = tf.Variable(5.0)

with tf.GradientTape() as tape:
    y = x**2

gradient = tape.gradient(y,x)

print(gradient)
```
