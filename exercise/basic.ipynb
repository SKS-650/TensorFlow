# Exercise 1: Tensor Creation and Properties
Problem 1

Create a scalar tensor with value 25.

Solution

```python
import tensorflow as tf

scalar = tf.constant(25)

print(scalar)
print("Shape:", scalar.shape)
print("Rank:", tf.rank(scalar))
```

# Problem 2

[10, 20, 30, 40, 50]

Print:

- Tensor
- Shape
- Rank  

Solution

```python
import tensorflow as tf

vector = tf.constant([10, 20, 30, 40, 50])

print(vector)
print("Shape:", vector.shape)
print("Rank:", tf.rank(vector))
```

# Problem 4

Convert an integer tensor to float32.

Solution

```python
import tensorflow as tf

x = tf.constant(
    [1,2,3],
    dtype=tf.int32
)

y = tf.cast(
    x,
    tf.float32
)

print(y)
```

# Exercise 4: Shape and Reshaping
### Problem 1

Create:

[1,2,3,4,5,6]

Convert into:

2 x 3

matrix.

Solution

```python
import tensorflow as tf

x = tf.constant([1,2,3,4,5,6])

y = tf.reshape(
    x,
    (2,3)
)

print(y)
```

# Problem 4

Expand dimension of:

[1,2,3]  

Solution

```python
import tensorflow as tf

x = tf.constant([1,2,3])

y = tf.expand_dims(
    x,
    axis=0
)

print(y)
```

# Problem 5

Remove extra dimension.

Solution

```python
import tensorflow as tf

x = tf.constant([
    [[1,2,3]]
])

y = tf.squeeze(x)

print(y)
```

# Problem 3

Matrix multiplication.

Matrix A:

1 2  
3 4

Matrix B:

5 6  
7 8  

Solution

```python
import tensorflow as tf

a = tf.constant([
    [1,2],
    [3,4]
])

b = tf.constant([
    [5,6],
    [7,8]
])

print(
    tf.matmul(a,b)
)
```
